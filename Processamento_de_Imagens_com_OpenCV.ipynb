## Baixando Assets necessários

```python
# Função que será responsável por baixar as imagens usadas no notebook
def baixar_e_extrair(url, save_path):
    print(f"Baixando imagens....", end="")

    # Baixando arquivo zip com urllib
    urlretrieve(url, save_path)

    try:
        # Extraindo arquivo zip usando o pacote zipfile.
        with ZipFile(save_path) as z:
            # Extraindo conteudos do arquivo no mesmo diretório.
            z.extractall(os.path.split(save_path)[0])

        print("Finalizado")

    except Exception as e:
        print("\nArquivo Inválido.", e)
```

```python
URL = r"https://www.dropbox.com/s/qhhlqcica1nvtaw/opencv_bootcamp_assets_NB1.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB1.zip")

# Realizar o download caso asset não estiver baixado.
if not os.path.exists(asset_zip_path):
    baixar_e_extrair(URL, asset_zip_path)
```

## Ler imagens usando OpenCV

Podemos ler vários tipos de imagens com OpenCV, desde imagens em cinza até imagens com canais alfa usando **`cv2.imread()`**


### Explicação da Sintaxe

**`cv2.imread( "Nome/Caminho do arquivo", parâmetros)`**

**`cv2`** Nome do módulo

**`imread`** Abreviação de "Image Read"

Os parâmetros utilizados nela são:
- 0: para ler a imagem em escala de cinza
- 1: para ler a imagem em cores
- -1: ler a imagem incluido canais alfa

```python
# Lendo uma imagem como uma escala de cinza com o parâmetro 0 após o nome do arquivo, seria lido em cores caso fosse 1
imagem_tabuleiro = cv2.imread("checkerboard_18x18.png", 0)

# Imprimindo na tela informações da imagem (Valores em 8-bits de 0 a 255)
print(imagem_tabuleiro)
```

## Mostrando imagens com matplotlib

```python
# Imprimindo na tela a imagem e o seu tipo de dado
print("Tamanho da imagem (X, Y):", imagem_tabuleiro.shape)
print("Tipo de dado da imagem:", imagem_tabuleiro.dtype)
```

```python
# Imprimindo imagem
plt.imshow(imagem_tabuleiro)
```

Como pode ver, a imagem foi mostrada em Roxo e Amarelo, isso acontece pois o matplotlib usa mapas de cores próprios para mostrar valores, não sendo necessáriamente escalas de cinza, dando ao programador a liberdade de setar os próprios mapas de cores


```python
# Mudando o mapa de cores para escalas de cinza com o parâmetro cmap(Color Map ou Mapa de Cores)
plt.imshow(imagem_tabuleiro, cmap='gray')
```

### O Que aconteceu com a imagem, por que ficou azul?
> Isso acontece pois o OpenCV usa um formato diferente de outras aplicações para guardar os canais de uma imagem, armazenando em formato BGR ao invés de RGB

> Para a imagem ser mostrada da forma correta, temos que inverter os canais da imagem, um exemplo de como fazer isso está na célula de código abaixo

```python
# Manipulando cores da imagem como se fosse um array multidimensional
# [:, :, ]: Seleciona todos os pixels em todas as linhas e colunas da imagem.
# [::-1]: Inverte a ordem dos canais de cores, -1 significa "comece do último elemento e vá até o primeiro, com passo -1"
logotipo_img_canais_invertidos = logotipo_imagem[:, :, ::-1]
plt.imshow(logotipo_img_canais_invertidos)
```

## Dividindo e mesclando Canais de Cores

**`cv2.split()`** Divide uma array multidimensional em várias arrays de canal único

**`cv2.merge()`** Mescla arrays de canal único em uma única array multidimensional

```python
# Carrega a imagem no formato BGR (azul, verde, vermelho)
imagem_NovaZelandia_bgr = cv2.imread("New_Zealand_Lake.jpg", 1)

# Divide a imagem em componentes B,G,R
# Cada uma dessas variáveis (b,g,r) representam uma array numpy bidimensional que contém a intensidade de pixels da imagem
b, g, r = cv2.split(imagem_NovaZelandia_bgr)

# Mostra os canais
# Função .subplot() usada para plotar várias imagens de uma vez
plt.figure(figsize=[20,5])

# Canal Vermelho da Imagem da Nova Zelândia
plt.subplot(141);plt.imshow(r, cmap='gray');plt.title("Canal Vermelho");

# Canal Verde da Imagem da Nova Zelândia
plt.subplot(142);plt.imshow(g, cmap='gray');plt.title("Canal Verde");

# Canal Azul da Imagem da Nova Zelândia
plt.subplot(143);plt.imshow(b, cmap='gray');plt.title("Canal Azul");

# Mescla os canais individuais em uma imagem B,G,R
imagem_rgb_mesclada = cv2.merge((b,g,r))

# Mostra a imagem resultante da mesclagem dos canais
plt.subplot(144);plt.imshow(imagem_rgb_mesclada[:, :, ::-1]);plt.title("Canais mesclados");
```

## Convertendo para diferentes espaços de cores

**`cv2.cvtColor()`** Função que converte imagem de um espaço de cor para outro (RGB para Escala de Cinza, BGR e vice versa)



### Sintaxe:

**`cv2.cvtColor(imagem, codigo)`**

**`imagem`** : A imagem de entrada que será convertida

**`codigo`** : Codigo do espaço de cores para conversão

**`Documentação de códigos de cores da OpenCV`**: https://docs.opencv.org/4.x/d8/d01/group__imgproc__color__conversions.html

#### Mudando de BGR para RGB

```python
# A biblioteca OpenCV guarda canais de cores em uma ordem diferente de outras aplicações (BGR)
imagem_NovaZelandia_rgb = cv2.cvtColor(imagem_NovaZelandia_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_NovaZelandia_rgb)
```

#### Mudando para espaço de cores HSV (Hue, Saturation, Value)

**`Hue`**: Representa a cor da imagem

**`Saturation`**: Representa a intensidade da cor

**`Value`**: Representa o valor, o quão clara ou escura a cor é

```python
# Convertendo a imagem em BGR para HSV
imagem_hsv = cv2.cvtColor(imagem_NovaZelandia_bgr, cv2.COLOR_BGR2HSV)

# Separando em três variaveis, uma para cada componente usando a função .split()
h,s,v = cv2.split(imagem_hsv)

# Mostrando os canais
# Função .subplot() usada para plotar várias imagens de uma vez
plt.figure(figsize=[20,5])

# Canal H da imagem da Nova Zelândia
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("Canal H");

# Canal S da imagem da Nova Zelândia
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title("Canal S");

# Canal V da imagem da Nova Zelândia
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title("Canal V");

# Mostra a imagem no espaço de cores original
plt.subplot(144);plt.imshow(imagem_NovaZelandia_rgb);plt.title("Imagem Original")
```

## Modificando canais individualmente

```python
# Modificando o canal H individualmente, aumentando seu valor em 10
h_novo = h+10
imagem_NovaZelandia_mesclada = cv2.merge((h_novo, s, v))
imagem_NovaZelandia_rgb = cv2.cvtColor(imagem_NovaZelandia_mesclada, cv2.COLOR_HSV2RGB)


# Mostrando os canais
# Função .subplot() usada para plotar várias imagens de uma vez
plt.figure(figsize=[20,5])

# Canal H da imagem da Nova Zelândia
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("Canal H");

# Canal S da imagem da Nova Zelândia
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title("Canal S");

# Canal V da imagem da Nova Zelândia
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title("Canal V");

# Mostra a imagem no espaço de cores original
plt.subplot(144);plt.imshow(imagem_NovaZelandia_rgb);plt.title("Modificada")
```

## Salvando imagens

**``cv2.imwrite()``**: Função responsável por salvar imagens em um formato especificado, porém apenas imagem de canal único ou de três canais (Formato BGR) podem ser salvas usando ela

### Explicação da Sintaxe:

**``cv2.imwrite(Nome/Caminho do arquivo, imagem[parametros])``**


**``Documentação da OpenCV``**: https://docs.opencv.org/3.4/d8/d6a/group__imgcodecs__flags.html

```python
# Salvando imagem
cv2.imwrite("New_Zealand_Lake_SAVED.png", imagem_NovaZelandia_bgr)

# Mostrando imagem
Image(filename="New_Zealand_Lake_SAVED.png")
```
